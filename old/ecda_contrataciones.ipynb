```python
import json
   
# Opening JSON file
f = open('contratacionesabiertas_bulk_paquetes_json/contratacionesabiertas_bulk_paquete1.json',)
   
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
data[0]
```

```python
# Function for flattening
# json

def flatten_json(y):
	out = {}

	def flatten(x, name=''):

		# If the Nested key-value
		# pair is of dict type
		if type(x) is dict:

			for a in x:
				flatten(x[a], name + a + '_')

		# If the Nested key-value
		# pair is of list type
		elif type(x) is list:

			i = 0

			for a in x:
				flatten(a, name + str(i) + '_')
				i += 1
		else:
			out[name[:-1]] = x

	flatten(y)
	return out
```

```python
flatten_json(data)
```

```python
data_flatten = [flatten_json(x) for x in data]
```
